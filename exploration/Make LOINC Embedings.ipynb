```python
import nltk
token_counts = {}
from nltk.stem import PorterStemmer
ps = PorterStemmer()

for r in loincs_as_dict:
    r["textTokens"] = set([ps.stem(v.lower()) for v in nltk.tokenize.word_tokenize(r['text'] )])
    r["relatedTokens"] = set([ps.stem(v.lower()) for v in nltk.tokenize.word_tokenize( r['RELATEDNAMES2'])])
    for t in r["textTokens"] | r["relatedTokens"]:
        if t not in token_counts:
            token_counts[t] = 0
        token_counts[t] += 1
len(token_counts)
```

```python
def find_hits(q):
    qtokens = set([ps.stem(v.lower()) for v in nltk.tokenize.word_tokenize(q)])
    print(qtokens)
    results = []
    for r in loincs_as_dict:
        matchText = r['textTokens'] & qtokens
        matchRelated = r['relatedTokens'] & qtokens
        #print(r['tokens'] & qtokens, len(r['tokens'] & qtokens))
        if len(matchText) + len(matchRelated) > 0:
            results.append([sum([10 * 60000/token_counts[t] for t in matchText]) + \
                                 sum([60000/token_counts[t] for t in matchRelated]), r])
    results.sort(key=lambda v: v[0], reverse=True)
    return results[:100]

for r in find_hits("blood glucose")[:50]:
    print(r[1]['LOINC_NUM'],r[1]['text'])#, f"(units: {r[1]['EXAMPLE_UCUM_UNITS']})")

len(loincs_as_dict)
#from txtai.embeddings import Embeddings
```

```python
import sqlite3

import csv

loincs_as_dict = []
# open the CSV file in read mode
with open('LoincTable/Loinc.csv', 'r') as csvfile:
    # create a CSV reader object
    csvreader = csv.reader(csvfile)
    header = next(csvreader)
    indexes = {header[i]: i for i in range(len(header))}
    print(indexes)
    for row in csvreader:
        loinc = {}
        #print(row)
        for key, index in indexes.items():
            loinc[key] = row[index]
        loinc["COMMON_TEST_RANK"] = int(loinc["COMMON_TEST_RANK"])
        if loinc["COMMON_TEST_RANK"] == 0:
            loinc["COMMON_TEST_RANK"] = 1000000
        loinc["text"] = loinc["LONG_COMMON_NAME"]
        loincs_as_dict.append(loinc)
    loincs_as_dict.sort(key=lambda l: l["COMMON_TEST_RANK"])


#embeddings = Embeddings({"path": "sentence-transformers/nli-mpnet-base-v2", "content": True})
#embeddings.index([(l["LOINC_NUM"], l, None) for l in loincs_as_dict])
#embeddings.save("loinc_full_longname")

# Connect to the database file
conn = sqlite3.connect("loinc2.db")

# Create a cursor object
cur = conn.cursor()
cur.execute("DROP TABLE IF EXISTS loinc")
cur.execute("CREATE VIRTUAL TABLE IF NOT EXISTS loinc USING fts5(code, display, related, common_test_rank, units, tokenize = porter)")
cur.execute("INSERT into loinc(loinc, rank) values('rank', ' bm25(0, 10, 10, 0)')")

for r in loincs_as_dict:
    code = r['LOINC_NUM']
    display = r['text']
    related = r['RELATEDNAMES2']
    common_test_rank = r['COMMON_TEST_RANK']
    units = r['EXAMPLE_UCUM_UNITS']
    cur.execute("INSERT INTO loinc (code, display, related, common_test_rank, units) VALUES (?, ?, ?, ?, ?)", (code, display, related, common_test_rank, units));

# Define a function to perform a full-text search on the table
def search_loinc(query, limit = 100, skip = 0):
  cur.execute("SELECT *, ROUND(rank, 0) AS rank FROM loinc WHERE loinc MATCH ? ORDER BY rank, length(display) limit ? offset ?", (" OR ".join(query.replace("-", " ").split(" ")), limit, skip,))
  results = cur.fetchall()
  print(f"---\nSearch results for '{query}':")
  return [f"{row[0]}, {row[1]} ctr {row[3]} {row[-1]}" for row in results]

# Perform some sample searches
#search_loinc("hcv antibody screening")
#search_loinc("cholesterol")
#search_loinc("blood")

# conn.close()
```

```python

import sqlite3
import random

# Connect to the database file
conn = sqlite3.connect("loinc.db")

# Create a cursor object
cur = conn.cursor()

# Define a function that randomly samples a LOINC code from the table
def sample_loinc():
    # Get the number of rows in the table
    cur.execute("SELECT COUNT(*) FROM loinc")
    n = cur.fetchone()[0]

    # Generate a random row number
    row = random.randint(1, n)

    # Get the LOINC code and other information from that row
    cur.execute("SELECT * FROM loinc WHERE rowid = ?", (row,))
    code, display, related, units = cur.fetchone()

    # Return the LOINC code and other information as a tuple
    return (code, display, related, units)

# Define a function that creates a prompt for a large language model to imagine 10 examples of clinical documentation phrasing for a LOINC code
def create_prompt(codes):
    prompt = f"Given the following LOINC code and its information, for each code please imagine 10 examples of english language clinical documentation phrasing that might refer to this LOINC concept in a realistic way.\n\n" + \
             "Never use the LOINC numeric code in your examples, and try to vary the phrasing so you're not directly using the same words as the LOINC display name every time."
    for [code, display, related, units] in codes:
        # Start the prompt with some instructions and context
        prompt += f"LOINC code: {code}\n"
        prompt += f"Display name: {display}\n"
        prompt += f"Related names: {related}\n"
        prompt += f"Units: {units}\n\n"

    # End the prompt with an empty line and a token to indicate where the model should generate text
    prompt += "\n<|endofprompt|>\n"

    # Return the prompt as a string
    return prompt

# Test the functions by sampling a LOINC code and creating a prompt for it
codes = [sample_loinc() for i in range(4)]
prompt = create_prompt(codes)
print(prompt)
```
